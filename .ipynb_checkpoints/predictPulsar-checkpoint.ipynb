## 1. Problem
Classify stars between Pulsar (1) and non-Pulsar (0).
### 1.1 Proposal
Test various Machine Learning (ML) algorithms to try to predict Pulsar stars. The dataset used is available at [1] and was taken from [2].
## 2. Data collection
The data was collected and prepared by Pavan Raj [2].
### 2.1 Dataset - [pulsar_stars.csv](https://raw.githubusercontent.com/spingegod/ProjetoTI_part1/master/ProjetoTI/pulsar_stars.csv) [1]

*According to [2], the existing data in the dataset are:*

**rows**: 16,259 spurious examples (RFI/noise), and 1,639 real pulsar examples. Total of 17,898 rows.  

**columns**: 9 features, namely:  
  1. Mean of the integrated profile.
  2. Standard deviation of the integrated profile.
  3. Excess kurtosis of the integrated profile.
  4. Skewness of the integrated profile.
  5. Mean of the DM-SNR curve.
  6. Standard deviation of the DM-SNR curve.
  7. Excess kurtosis of the DM-SNR curve.
  8. Skewness of the DM-SNR curve.
  9. Class (1=Pulsar, 0=non-Pulsar)

So, the features are columns 1 to 8, and the label is the column 9.

# 3. Data Preparation

```python
# Import needed libraries
import pandas as pd
# import pandas_profiling complete report
import numpy as np
import datetime
import warnings
import matplotlib.pyplot as plt
%matplotlib inline
warnings.simplefilter(action='ignore')

from sklearn.metrics import mean_absolute_error, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

current_date: str = datetime.date.today().strftime("%y-%m-%d")
```

## 3.1 Download dataset

```python
def saveData(df, folder='', name='temp'):
    """Save a dataframe to a csv file"""
    if folder == 'raw':
        folder = 'data/raw/'
    elif folder == 'processed':
        folder = 'data/processed/'
    elif folder == 'results':
        folder = 'results/'
    return df.to_csv(folder + str(current_date) + '_' + name + '.csv')


# dataset developed by [2]
url_dataset = 'https://raw.githubusercontent.com/spingegod/ProjetoTI_part1/master/ProjetoTI/pulsar_stars.csv'

data = pd.read_csv(url_dataset)

# save raw data
saveData(data, 'raw', 'pulsar_stars.csv')
```

## 3.3 Split data in train/test

```python
X = data.drop(['target_class'], axis=1)
y = data.drop(data.columns[0:8], axis=1)
```

```python
# 80/20
X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=0.8, random_state=0, stratify=y)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## 3.4 Normalization

Note: The data is unbalanced, to solve this problem the **undersampling** technique was chosen. For this, I will decrease the negative samples so that they are the same size as the positive ones.

```python
# join train dataset
train_data, train_data['target_class'] = X_train, y_train

train_data.shape
```

```python
# split positives and negatives data
dataPositives = train_data.loc[train_data['target_class'] == 1]
dataNegatives = train_data.loc[train_data['target_class'] == 0].sample(
    n=dataPositives.shape[0], random_state=0)

# concatenate both datasets
normalizedData = pd.concat([dataPositives, dataNegatives])
```

```python
# looking at the data again
normalizedNegatives = normalizedData.loc[:, 'target_class'].isin([0]).sum()
normalizedPositives = normalizedData.loc[:, 'target_class'].isin([1]).sum()

print('non-Pulsar: {} \nPulsar: {}'.format(
    normalizedNegatives, normalizedPositives))
```

```python
# split data into X and y
X_norm = normalizedData.drop(['target_class'], axis=1)
y_norm = normalizedData['target_class']

X_norm.shape, y_norm.shape
```
